```python
#1. общий анализ каждого столбца
for w in stud_math.columns:
    display(pd.DataFrame(stud_math[w].value_counts()))
    display(stud_math[w].describe())
    
    # Для беглой оценки влияния показателей на оценку добавлю скрипичную диаграмму
    plt.figure(figsize=(13,5), dpi= 100)
    sns.violinplot(x=w, y='score', data=stud_math, scale='width', inner='quartile')
    
    # Decoration
    plt.title(comment[w], fontsize=15)
    plt.show()
```

```python
# Выделяю столбцы с числовыми переменными для устранения выбросов.
#'age', 'Fedu', 'famrel', 'absences', 'score'
#Вместо выбросов ставлю None.

for w in stud_math[{'age', 'Fedu', 'absences'}].columns:
    
    median = stud_math[w].median()
    IQR = stud_math[w].quantile(0.75) - stud_math[w].quantile(0.25)
    perc25 = stud_math[w].quantile(0.25)
    perc75 = stud_math[w].quantile(0.75)
    print('\033[92m','***',w,'***')
    print('25-й перцентиль: {},'.format(perc25), '75-й перцентиль: {},'.format(perc75)
          , "IQR: {}, ".format(IQR),"Границы выбросов: [{f}, {l}].".format(f=perc25 - 1.5*IQR, l=perc75 + 1.5*IQR))
    f=perc25 - 1.5*IQR
    l=perc75 + 1.5*IQR
    stud_math[w] = stud_math[w].apply(lambda x: None if x < f or x > l else x)
    
    display(stud_math[w].value_counts())
    display(stud_math[w].describe())
```

```python
#Анализ номинативных переменных

def get_boxplot(column):
    fig, ax = plt.subplots(figsize = (14, 4))
    sns.boxplot(x=column, y='score', 
                data=stud_math.loc[stud_math.loc[:, column].isin(stud_math.loc[:, column].value_counts().index[:10])],
               ax=ax)
    plt.xticks(rotation=45)
    ax.set_title('Boxplot for ' + column)
    plt.show()
```

```python
coll=['school', 'sex', 'address', 'famsize', 'Pstatus', 'Medu', 'Fedu',
       'Mjob', 'Fjob', 'reason', 'guardian', 'traveltime', 'studytime',
       'failures', 'schoolsup', 'famsup', 'paid', 'activities', 'nursery',
       'studytime, granular', 'higher', 'internet', 'romantic', 'famrel',
       'freetime', 'goout', 'health']
```

```python

for col in coll:
    get_boxplot(col)
```

```python
# Проверим, есть ли статистическая разница в распределении оценок по номинативным признакам, 
# с помощью теста Стьюдента. Проверим нулевую гипотезу о том, что распределения оценок 
# по различным параметрам неразличимы:
def get_stat_dif(column):
    cols = stud_math.loc[:, column].value_counts().index[:10]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(stud_math.loc[stud_math.loc[:, column] == comb[0], 'score'], 
                        stud_math.loc[stud_math.loc[:, column] == comb[1], 'score']).pvalue \
            <= 0.05/len(combinations_all): # Учли поправку Бонферони
            print('Найдены статистически значимые различия для колонки', column)
            break
```

```python
for col in coll:
    get_stat_dif(col)
```

```python
# Добавлю дополнительный столбец zone, где зоне риска будет соответствовать 0, иначе 1.
r = 55
stud_math['zone']=stud_math.score.apply(lambda x: 0 if x < r else 1)
risk=len(stud_math.loc[stud_math['zone']==0])
pr=risk/len(stud_math)*100
print('Студентов в зоне риска:', risk, '('+ str(round(pr,1))+'%) от общей выборки.')
display(stud_math)
```

```python
#   Анализирую оставшиеся параметры
xr=['address', 'Medu', 'Fedu', 'Mjob', 'Fjob','studytime', 'failures', 'schoolsup', 
'studytime, granular', 'goout', 'absences']
```

```python
# Разделю датафрейм по зонам. Найду доли самых влиятельных показателей. Сопоставлю эти доли по зонам 0 и 1.
col=[]   # имя столбца
name=[]  # имя показателя внутри столбца
c1=[]    # значение самой весомой доли показателя в зоне риска
c2=[]    # значение самой весомой доли показателя в зоне успеха

zone0=stud_math.loc[stud_math.zone==0]
zone1=stud_math.loc[stud_math.zone==1]
for w in xr:
    
    #z0=pd.DataFrame(zone0[w].value_counts(normalize=True))
    #z1=pd.DataFrame(zone1[w].value_counts(normalize=True))
    
    z0=pd.DataFrame(zone0[w].value_counts())
    z1=pd.DataFrame(zone1[w].value_counts())
    c1.append(round(z0.iloc[0][0],2))
    c2.append(round(z1.iloc[0][0],2))
    name.append(z0.index[0])
    col.append((z0.columns[0]))

d={'name':pd.Series(name, index=col),'z0':pd.Series(c1, index=col), 'z1':pd.Series(c2, index=col)}
df1 = pd.DataFrame(d)
df1['k']=round((d['z0']/d['z1']),2) #доля студентов зоны риска от зоны успеха
display(df1)
```

```python
Общие выводы
По данной выборке в зоне риска нет преобладания ни одного из показателей.
```

```python

col=[] # для названия столбца
name=[] # для названия показателя внутри столбца
k=[] # коэффициент корреляции

xr=['address', 'Medu', 'Fedu', 'Mjob', 'Fjob','studytime', 'failures', 'schoolsup', 
'goout']
study=stud_math[xr]
for w in study.columns:
    display(pd.DataFrame(study[w].value_counts()))
    
    # Далее сделаю сравнение серий с разделением по зонам.
     
    #z0=pd.DataFrame(stud_math[w].loc[stud_math['zone']==0].value_counts(normalize=True))
    #z1=pd.DataFrame(stud_math[w].loc[stud_math['zone']==1].value_counts(normalize=True))
    #d=pd.concat([round(z0*100,0), round(z1*100,0)], axis=1, sort=False)
    
    z0=pd.DataFrame(stud_math[w].loc[stud_math['zone']==0].value_counts())
    z1=pd.DataFrame(stud_math[w].loc[stud_math['zone']==1].value_counts())
    d=pd.concat([z0, z1], axis=1, sort=False)

    d.columns = ['z0', 'z1']
    
    # Добавлю столбец, показывающий доли показателей зоны риска и зоны успеха.
    
    d['k']=round((d['z0']/(d['z1']+d['z0'])),2)

    display(d)
    
    
    # Соберу в отдельные списки показатели с высокими коэффициентами 'k'
    dk=d.loc[d['k']==d['k'].max()]
    dk['indicator']=w
    display(dk)
    a=dk.values.tolist()[0]
    b=dk.index
    a.insert(0,b[0])
   # k.extend(b)
    k.append(a)
   
    
    # Для дополнительной оценки влияния показателей на попадание в зону риска добавлю скрипичную диаграмму
    plt.figure(figsize=(13,5), dpi= 100)
    sns.violinplot(x=w, y='zone', data=stud_math, scale='width', inner='quartile')

    # Decoration
    plt.title(comment[w], fontsize=15)
    plt.show()
```

```python
result=pd.DataFrame(k[0:9])
result.columns=['name', 'z0', 'z1', 'k', 'indicator']
# Оставляю 7 самых влиятельных на зону риска показателя, сортировка по степени значимости.
display(result.sort_values(by=['k'], ascending=False)[0:7])
```
